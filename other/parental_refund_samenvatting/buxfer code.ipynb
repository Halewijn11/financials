# CSV to excel

```python
import numpy as np
import pandas as pd
import os


def convert_to_excel(filename): 
    filepath = r'C:\Users\User\OneDrive\Documenten\financials\monthly expenses'
    filename = filename + '.csv'
    df = pd.read_csv(filepath + '/' + filename)
    df.loc[:, 'abs'] = df.loc[:, 'Amount'].abs()
    df =df.sort_values('abs')
    df = df.drop('abs', axis = 1)
    df = df.drop('IOU', axis = 1)
    df = df.drop('Memo', axis =1)
    df.to_excel(filepath + '/' + filename[:-3] + 'xlsx', index = False)
    tot = df.loc[:, 'Amount'].sum()
    display(str(tot))
```

```python
filename = input('What is the filename?')
convert_to_excel(filename)
```
